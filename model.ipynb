```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df=df[df['overs']>5.0]
df.head()
```

```python
from datetime import datetime as dt
df['date']=df['date'].apply(lambda x:dt.strptime(x,'%Y-%m-%d'))
```

```python
df['bat_team'].unique()
```

```python
main_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df=df[df['bat_team'].isin(main_teams) & df['bowl_team'].isin(main_teams)]
df.head()
```

```python
df.drop(['mid','batsman','bowler','striker','non-striker'],axis=1,inplace=True)
df.head()
```

```python
df.head()
```

```python
df.dtypes
```

```python
final_df=df.drop(['venue'],axis=1)
```

```python
final_df.head()
```

```python
max(df['date'].dt.year)
```

```python
final_df=pd.get_dummies(final_df)
```

```python
y_train=final_df[final_df['date'].dt.year <=2016 ]['total']
y_test= final_df[final_df['date'].dt.year >2016 ]['total']
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
X_train=final_df.drop(['total'],axis=1)[final_df['date'].dt.year <=2016].drop(['date'],axis=1)
X_train.head()
```

```python
X_test=final_df.drop(['total'],axis=1)[final_df['date'].dt.year > 2016].drop(['date'],axis=1)
X_test.head()
```

```python
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
regressor=RandomForestRegressor()
```

```python
n_estimators=[50,100,150,200,250]
max_depth=[5,10,15,20,25,30,35]
max_features=['auto', 'sqrt']
min_samples_split=[2, 5, 10, 15, 100]
min_samples_leaf=[1, 2, 5, 10]
```

```python
parameters={
     'n_estimators':n_estimators,
     'max_depth':max_depth,
     'max_features':max_features,
     'min_samples_leaf':min_samples_leaf,
     'min_samples_split':min_samples_split
}
```

```python
rf=RandomizedSearchCV(regressor,parameters,scoring='neg_mean_squared_error',cv=5,n_jobs=1)
```

```python
rf.fit(X_train,y_train)
```

```python
pred=rf.predict(X_test)
```

```python
import seaborn as sns
sns.distplot(y_test-pred)
```

```python
import pickle
file=open('model.pkl','wb')
pickle.dump(rf,file)
```
